```python
# Extraccion de datos del TOP1000 de Twitchtracker

data_results=[]

def extrac_top1000():
    #numerador_pagina = 1
    #url = 'https://twitchtracker.com/channels/ranking?page=' + str(numerador_pagina)
    numerador_pagina = 0
    url = 'https://twitchtracker.com/subscribers?page='
    
    #El numero de range(XX) marca el numerador final de la pagina 50 serian el Top1000
    for i in range(50):
        # Extraccion de datos2 del TOP50 de Twitchtracker
        
        numerador_pagina = int(numerador_pagina) + 1
        
        url_pagina = url + str(numerador_pagina)
        
        driver.get(url_pagina)
        
        # Nombre del Streamer
        nombres_streamer = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[4]')
        # Current Active Subs
        current_active_subs = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[5]')
        # Sub type Paid
        subs_type_paid = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[6]')
        # Sub type Prime
        subs_type_prime = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[7]')
        # Sub type Gifted
        subs_type_gifted = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[8]')
        # Sub type Tier1
        subs_type_tier_one = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[9]')
        # Sub type Tier2
        subs_type_tier_two = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[10]')
        # Sub type Tier3
        subs_type_tier_three = driver.find_elements(By.XPATH, '//*[@id="channels"]/tbody/tr/td[11]')


        for item in range(len(nombres_streamer)):
            temporal_data={
                            #'Numero': numero[item].text,
                            #'URL imagen Streamer': imagen[item].get_attribute('innerHTML'),
                            'Nombres Streamers': nombres_streamer[item].text,
                            'Current Active Subs': current_active_subs[item].text,
                            'Subs Type Paid': subs_type_paid[item].text,
                            'Subs Type Prime': subs_type_prime[item].text,
                            'Subs Type Gifted': subs_type_gifted[item].text,
                            'Subs Type Tier 1($4.99)': subs_type_tier_one[item].text,
                            'Subs Type Tier 2($9.99)': subs_type_tier_two[item].text,
                            'Subs Type Tier 3($24.99)': subs_type_tier_three[item].text,
                            #'Raw data': raw_data[item].text,
                            }
            data_results.append(temporal_data)
    
extrac_top1000()
```
